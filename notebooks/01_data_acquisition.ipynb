```python
# La Liga 2015/2016 — get all matches
matches = sb.matches(competition_id=11, season_id=27)
print("Number of matches:", matches.shape[0])
matches[['match_id', 'home_team', 'away_team', 'home_score', 'away_score']].head(10)
```

```python
import pandas as pd

all_shots = []                       # empty pile to collect shots
match_ids = matches['match_id'].tolist()

for i, mid in enumerate(match_ids):
    try:
        events = sb.events(match_id=mid)          # get everything in the match
        shots = events[events['type'] == 'Shot']  # keep only shots
        all_shots.append(shots)
    except Exception as e:
        print(f"Skipped match {mid}: {e}")

    if (i + 1) % 50 == 0:                          # progress every 50 matches
        print(f"Processed {i + 1}/{len(match_ids)} matches...")

# Stack all the shots into one big table
shots_df = pd.concat(all_shots, ignore_index=True)
print("\nDone! Total shots collected:", shots_df.shape[0])
```

```python
# Look at the key columns for building xG
shots_df[['player', 'team', 'shot_statsbomb_xg', 'shot_body_part',
          'shot_technique', 'shot_type', 'shot_outcome', 'location']].head(10)
```

```python
# How do shots end up? (our target variable)
print(shots_df['shot_outcome'].value_counts())
print()

# Turn outcome into a simple 1 = goal, 0 = not goal
shots_df['goal'] = (shots_df['shot_outcome'] == 'Goal').astype(int)
print("Goal rate:", round(shots_df['goal'].mean() * 100, 1), "%")
```

```python
# Does StatsBomb's xG roughly match the real goal rate? (sanity check)
print("Actual goal rate:      ", round(shots_df['goal'].mean() * 100, 1), "%")
print("Average StatsBomb xG:  ", round(shots_df['shot_statsbomb_xg'].mean() * 100, 1), "%")
```

```python
import sqlite3

# Create/connect to a database file in your project folder
conn = sqlite3.connect('football_xg.db')

# Keep the columns we need
cols_to_save = ['id', 'match_id', 'player', 'team', 'shot_statsbomb_xg',
                'shot_body_part', 'shot_technique', 'shot_type',
                'shot_outcome', 'location', 'goal']

shots_to_save = shots_df[cols_to_save].copy()
# location is a list -> store as text so SQLite accepts it
shots_to_save['location'] = shots_to_save['location'].astype(str)

shots_to_save.to_sql('shots', conn, if_exists='replace', index=False)
print("Saved to database!")

# Prove it with an actual SQL query
query = "SELECT COUNT(*) AS total_shots, SUM(goal) AS total_goals FROM shots"
result = pd.read_sql(query, conn)
print(result)
```
